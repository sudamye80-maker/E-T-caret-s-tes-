```python
import random

def loto_cekilisi_yap(adet=6, baslangic=1, bitis=49):
    # sample fonksiyonu belirtilen aralıkta tekrarsız rastgele sayılar seçer
    cekilen_sayilar = random.sample(range(baslangic, bitis + 1), adet)
    
    # Sayıları küçükten büyüğe sıralar
    cekilen_sayilar.sort()
    
    return cekilen_sayilar

# Çekilişi çalıştır ve ekrana yazdır
sonuc = loto_cekilisi_yap()
print(f"Şanslı Numaralar: {sonuc}")
```

```python
import random

def loto_sayi_ureteci(max_sayi, adet):
    """Belirtilen aralıkta rastgele loto sayıları üretir."""
    # sample() fonksiyonu ile sayılar birbirinden farklı ve tekrarsız seçilir
    return sorted(random.sample(range(1, max_sayi + 1), adet))

# 1-80 arası On Numara için 22 adet rastgele tahmin oluşturalım
cekilis = loto_sayi_ureteci(80, 22)

print(f"Çekilen Sayılar: {cekilis}")

# İsteğe bağlı filtreleme veya boş tahmin kontrolü
def bos_tahmin_filtresi(uretilen_liste, tahminleriniz):
    # Doğru tahmin edilen sayıları bul
    dogru_bilinenler = set(uretilen_liste).intersection(tahminleriniz)
    
    if len(dogru_bilinenler) == 0:
        print("Tebrikler! Hiçbir sayıyı bilmediniz (Boş Tahmin).")
    else:
        print(f"Bilenen Sayılar: {dogru_bilinenler}")

# Örneğin sizin tahminleriniz:
kullanici_tahminleri = {5, 12, 28, 45, 60, 75}
bos_tahmin_filtresi(cekilis, kullanici_tahminleri)
```

```python
yapilacaklar = []

while True:
    print("\n--- YAPILACAKLAR LİSTESİ ---")
    print("1. Görevleri Göster")
    print("2. Görev Ekle")
    print("3. Görev Sil")
    print("4. Çıkış")
    
    secim = input("Lütfen bir işlem seçin (1-4): ")
    
    if secim == "1":
        if not yapilacaklar:
            print("Listeniz boş.")
        else:
            print("\nGörevleriniz:")
            for index, gorev in enumerate(yapilacaklar, 1):
                print(f"{index}. {gorev}")
                
    elif secim == "2":
        yeni_gorev = input("Eklenecek görevi yazın: ")
        yapilacaklar.append(yeni_gorev)
        print(f"'{yeni_gorev}' listeye eklendi.")
        
    elif secim == "3":
        if not yapilacaklar:
            print("Silinecek görev yok.")
        else:
            print("\nGörevleriniz:")
            for index, gorev in enumerate(yapilacaklar, 1):
                print(f"{index}. {gorev}")
            silinecek_no = int(input("Silmek istediğiniz görevin numarasını girin: "))
            if 0 < silinecek_no <= len(yapilacaklar):
                silinen = yapilacaklar.pop(silinecek_no - 1)
                print(f"'{silinen}' listeden silindi.")
            else:
                print("Geçersiz görev numarası.")
                
    elif secim == "4":
        print("Programdan çıkılıyor. İyi günler!")
        break
    else:
        print("Geçersiz seçim. Lütfen 1, 2, 3 veya 4 yazın.")
```

```python
# Öğrenci notlarını listeleyelim
notlar = [45, 60, 78, 90, 85, 30, 95, 65]

# Temel Analiz Fonksiyonu
def not_dokum_analizi(not_listesi):
    toplam = sum(not_listesi)
    adet = len(not_listesi)
    
    # İstatistikler
    ortalama = toplam / adet
    en_yuksek = max(not_listesi)
    en_dusuk = min(not_listesi)
    
    # Geçme kalma analizi (50 baraj puanı)
    gecenler = [n for n in not_listesi if n >= 60]
    kalanlar = [n for n in not_listesi if n < 60]
    gecen_sayisi = len(gecenler)
    kalan_sayisi = len(kalanlar)

    print("--- Not Döküm Analizi ---")
    print(f"Toplam Öğrenci Sayısı: {adet}")
    print(f"Sınıf Ortalaması: {ortalama:.2f}")
    print(f"En Yüksek Not: {en_yuksek}")
    print(f"En Düşük Not: {en_dusuk}")
    print(f"Geçen Öğrenci Sayısı: {gecen_sayisi}")
    print(f"Kalan Öğrenci Sayısı: {kalan_sayisi}")

# Fonksiyonu çalıştırıyoruz
not_dokum_analizi(notlar)
```

```python
ogrenci_sistemi = {}

def ogrenci_ekle(isim):
    ogrenci_sistemi[isim] = {"vize": 0, "final": 0, "ortalama": 0.0}
    print(f"{isim} sisteme eklendi.")

def not_gir(isim, vize_notu, final_notu):
    if isim in ogrenci_sistemi:
        ogrenci_sistemi[isim]["vize"] = vize_notu
        ogrenci_sistemi[isim]["final"] = final_notu
        # Ağırlıklı ortalama hesaplama (Vize %40, Final %60)
        ortalama = (vize_notu * 0.4) + (final_notu * 0.6)
        ogrenci_sistemi[isim]["ortalama"] = ortalama
        print(f"{isim} için notlar kaydedildi. Ortalama: {ortalama}")
    else:
        print("Öğrenci bulunamadı!")

def sistem_ozeti():
    print("\n--- Öğrenci Not Listesi ---")
    for isim, bilgiler in ogrenci_sistemi.items():
        durum = "Geçti" if bilgiler["ortalama"] >= 50 else "Kaldı"
        print(f"Öğrenci: {isim} | Ortalama: {bilgiler['ortalama']} | Durum: {durum}")

# Örnek Kullanım:
ogrenci_ekle("Ayşe")
not_gir("Ayşe", 45, 60)
sistem_ozeti()
```
